```python
import nltk
import random
from nltk.corpus import movie_reviews
```

Using the movie reviws dataset to classify them into positive or negative review

```python
movie_reviews.categories()
```

```python
movie_reviews.fileids('neg')
```

```python
documents = [(list(movie_reviews.words(fileid)), category)
             for category in movie_reviews.categories()
             for fileid in movie_reviews.fileids(category)]

random.shuffle(documents)

print(documents[1])#where the first element is a list the words, and the 2nd element is the "pos" or "neg" label.
```

```python
all_words = []
for w in movie_reviews.words():
    all_words.append(w.lower())

all_words = nltk.FreqDist(all_words)
print(all_words.most_common(15))
print(all_words["stupid"])
```

```python
type(all_words)
```

```python
word_features = list(all_words.keys())[:3000]
word_features
```

```python
def find_features(document):
    words = set(document)
    features = {}
    for w in word_features:
        features[w] = (w in words)

    return features
```

```python
featuresets = [(find_features(rev), category) for (rev, category) in documents]
len(featuresets)
```

```python
#Splitting the data into train and test
training_set = featuresets[:1900]
testing_set = featuresets[1900:]
```

###  Classification using NaiveBayes

```python
classifier = nltk.NaiveBayesClassifier.train(training_set)
```

```python
print("Classifier accuracy percent:",(nltk.classify.accuracy(classifier, testing_set))*100)
```

```python
#the most valuable words are when it comes to positive or negative reviews:
classifier.show_most_informative_features(15)
```

What this tells you is the ratio of occurences in negative to positive, or visa versa, for every word.

### Saving the Classifier
Instead of training the classifier everytime we wan tto use it, we should rather save the classifier using the pickle module. 
<br> For that, we will open a pickle file in 'wb' mode, dump the classifier there and close the file.
<br> To retrieve, open the file in 'rb' mode, load the classifier and close.
***
<div class="alert alert-block alert-warning">
<b>Warning:</b> The pickle module is not secure.<b>Only unpickle data you trust.</b>
It is possible to construct malicious pickle data which will execute arbitrary code during unpickling. Never unpickle data that could have come from an untrusted source, or that could have been tampered with.
Consider signing data with hmac if you need to ensure that it has not been tampered with.
Safer serialization formats such as json may be more appropriate if you are processing untrusted data.
</div>
<br> For more details on pickle refer <a href="https://docs.python.org/3/library/pickle.html">[Pickle Documentation]</a>

```python
import pickle
```

```python
save_classifier = open("naivebayes_nlp_txt_clsf.pickle","wb")
pickle.dump(classifier, save_classifier)
save_classifier.close()
```

```python
classifier_f = open("naivebayes_nlp_txt_clsf.pickle", "rb")
classifier = pickle.load(classifier_f)
classifier_f.close()
```
